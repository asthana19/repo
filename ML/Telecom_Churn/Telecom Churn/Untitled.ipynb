```python
import warnings
warnings.filterwarnings('ignore')
import numpy as np
import pandas as pd
```

```python
churn = pd.read_csv("churn_data.csv")
```

```python
customer = pd.read_csv("customer_data.csv")
```

```python
internet = pd.read_csv("internet_data.csv")
```

```python
df = pd.merge(churn, customer, how="inner", on = "customerID")
```

```python
df.head()
```

```python
telco = pd.merge(df, internet, how="inner", on = "customerID")
```

```python
telco.shape
```

```python
telco.info()
```

```python
# List of variables to map

varlist =  ['PhoneService', 'PaperlessBilling', 'Churn', 'Partner', 'Dependents']

# Defining the map function
def binary_map(x):
    return x.map({'Yes': 1, "No": 0})

# Applying the function to the housing list
telco[varlist] = telco[varlist].apply(binary_map)
```

```python
telco.head()
```

```python
# Let's drop the first column from status df using 'drop_first = True'

status = pd.get_dummies(telco[['Contract','PaymentMethod','gender','InternetService']], drop_first = True)
```

```python
status.head()
```

```python
telco = pd.concat([telco, status], axis = 1)
```

```python
telco.head()
```

```python
telco['MultipleLines'].astype('category').value_counts()
```

```python
ml = pd.get_dummies(telco[['MultipleLines']], prefix = 'MultipleLines')
```

```python
ml1= ml.drop('MultipleLines_No phone service',1)
```

```python
telco = pd.concat([telco, ml1], axis = 1)
```
